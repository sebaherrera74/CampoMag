```python
'''Leer un acrhivo .csv '''

import numpy as np
import matplotlib.pyplot as plt
import os
from datetime import datetime
from scipy.fft import fft, fftfreq

filename = '/content/25042025P-FET1.csv'
raw_data = open(filename)

data = np.loadtxt(raw_data, delimiter=",",skiprows=11)
#print(type (data))

#print(data.shape)                  #forma de los datos
fila,columna=data.shape            #Saco filas y columnas
print(fila,columna)
print(data.size)                   #Cantidad de filas datos en filas xcolumnas


#print(data[0:10000,0])
x=data[0:fila,6]
#print(x)

#print(data[0:10000,1])
y=data[0:fila,7]

# 3. Crear carpeta para guardar datos
fecha = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
seba = f"Medicion_{fecha}"
os.makedirs(seba, exist_ok=True)

#Grafica de Voltaje en el tiempo
plt.title('Id Vs Vds, Vrg=-2 Volt')
plt.xlabel('Vds')
plt.ylabel('Id')
plt.plot(x,y)
fig_path = os.path.join(seba, "curva_iv.png")
plt.savefig(fig_path)
plt.show()
plt.show()
```
